```python
import csv
Sentences = []
with open('/Users/keshen/PycharmProjects/sk/ISI/Malaysia/Day2/HandsOn-Amazon-GoogleProducts/GoogleProducts.csv', 'r') as csvfile:
    reader = csv.DictReader(csvfile)
    
    # Iterate over each row in the dataset
    for row in reader:
        # Extract relevant information from the row
        id = row['id']
        name = row['name']
        price = row['price']
        manufacturer = row['manufacturer']
        description = row['description']
        
        # Create a coherent sentence using string formatting
        sentence = f"The product with the ID {id} is named {name}. This product, priced at {price} and manufactured by {manufacturer}, is described as {description}."
        
        # Print or store the sentence for further processing
        print(sentence)
        Sentences.append(sentence)
```

```python
import spacy
import textacy
nlp = spacy.load('en_core_web_trf')

Entities = []
Relations = []
for text in Sentences:
    doc = nlp(text)
    E = []
    R = []
    for entity in doc.ents:
        entity_text = entity.text
        entity_type = entity.label_
        E.append((entity_type, entity_text))
        print("Entity extraction: ", entity_text, " | ", entity_type)

    for token in doc:
        if token.dep_ == 'ROOT':
            root = token
            relationships = []
            for child in token.children:
                relationships.append((child.text, child.dep_))  # Only parse the dependency of tokens in the sentence, but did not use any pattern templates to match the relations in sentences.
    
    """Use textacy to extract relations!"""
    tri = textacy.extract.triples.subject_verb_object_triples(doc)
    for item in tri:
        R.append(" ".join([str(x) for x in item[0]])+" -> "+" ".join([str(x) for x in item[1]])+ " -> "+" ".join([str(x) for x in item[2]]))
        print("Relation extraction: ", " ".join([str(x) for x in item[0]]), "->", " ".join([str(x) for x in item[1]]), "->", " ".join([str(x) for x in item[2]]))
    print("----------")
    Entities.append(E)
    Relations.append(R)
print(Relations[:5])
```

```python
import random

assert len(Sentences) == len(Relations) == len(Entities), print("Error")
    
# randomIdx = random.sample([i for i in range(len(Sentences))], 20)
randomIdx = [24, 41, 102, 104, 45, 57, 124, 120, 0, 39, 2, 125, 101, 63, 95, 19, 115, 14, 35, 91]
SampleEntities = [Entities[x] for x in randomIdx]
SampleRelations = [Relations[x] for x in randomIdx]
SampleSentences = [Sentences[x] for x in randomIdx]
for i in range(len(SampleSentences)):
    print(SampleSentences[i])
    print(SampleEntities[i])
    print(SampleRelations[i])
    print('--------')
```
